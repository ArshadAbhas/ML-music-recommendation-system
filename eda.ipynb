```python
numeric_feature = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_feature = [i for i in df.columns if df[i].dtype == 'O']
print("Categorical features:",categorical_feature)
print("numeric features:",numeric_feature)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Determine the number of rows and columns needed for the subplots
num_features = len(numeric_feature)
cols = 2  # Fixed number of columns
rows = math.ceil(num_features / cols)  # Calculate rows based on total features

plt.figure(figsize=(15, 5 * rows))  # Adjust height dynamically
plt.suptitle("Univariate Analysis", fontsize=16)

for i in range(num_features):
    plt.subplot(rows, cols, i + 1)
    sns.kdeplot(x=df[numeric_feature[i]], color='blue')
    plt.xlabel(numeric_feature[i])

# Adjust layout after all subplots are added
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the title
plt.show()
```

```python
#categorical Features univariate analysis

plt.figure(figsize=(40, 40))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(categorical_feature)):
    plt.subplot(3, 3, i+1)
    sns.countplot(x=df[categorical_feature[i]])
    plt.xlabel(categorical_feature[i])
    plt.tight_layout()
```
